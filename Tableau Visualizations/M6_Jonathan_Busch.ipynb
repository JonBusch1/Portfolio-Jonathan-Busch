```python
categories = ['Programming', 'Critical Thinking', 'Analytical Skills', 'Communication','Stupidity','Cringe']
categories = [*categories, categories[0]]
My_ability = [1, 1, 1, 1, 5,5]
Others = [3,3,4,3,3,3]
RCOBA = [2,3,3,4,3,3]
My_ability = [*My_ability, My_ability[0]]
Others = [*Others, Others[0]]
RCOBA = [*RCOBA, RCOBA[0]]
label_loc = np.linspace(start=0, stop=2 * np.pi, num=len(My_ability))
plt.figure(figsize=(4, 4))
plt.subplot(polar=True)
plt.plot(label_loc, My_ability, label='My Abilities')
plt.plot(label_loc, Others, label='MSDS Average')
plt.plot(label_loc, RCOBA, label='RCOBA Average')
plt.title('My Abilities vs. Others')
lines, labels = plt.thetagrids(np.degrees(label_loc), labels=categories)
plt.legend(loc='center')
plt.show()
```
